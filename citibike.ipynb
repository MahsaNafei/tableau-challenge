```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files_23 = ['Resources/202301.csv', 'Resources/202302.csv', 'Resources/202303.csv', 'Resources/202304.csv',
            'Resources/202305.csv', 'Resources/202306.csv', 'Resources/202307.csv', 'Resources/202308.csv', 
            'Resources/202309.csv', 'Resources/202310.csv', 'Resources/202311.csv', 'Resources/202312.csv']

# Create an empty list to store each dataframe
df_list_23 = []

# Loop through each file in the list
for file in files_23:
    # Read the file into a dataframe
    df_23 = pd.read_csv(file)

    # Append the dataframe to the list
    df_list_23.append(df_23)

# Concatenate all dataframes in the list into one dataframe
df_23 = pd.concat(df_list_23)

# Show dataframe
df_23.head()
```

```python
# List of file names to be combined 
files_22 = ['Resources/202201.csv', 'Resources/202202.csv', 'Resources/202203.csv', 'Resources/202204.csv',
            'Resources/202205.csv', 'Resources/202206.csv', 'Resources/202207.csv', 'Resources/202208.csv', 
            'Resources/202209.csv', 'Resources/202210.csv', 'Resources/202211.csv', 'Resources/202212.csv']

# Create an empty list to store each dataframe
df_list_22 = []

# Loop through each file in the list
for file in files_22:
    # Read the file into a dataframe
    df_22 = pd.read_csv(file)

    # Append the dataframe to the list
    df_list_22.append(df_22)

# Concatenate all dataframes in the list into one dataframe
df_22 = pd.concat(df_list_22)

# Show dataframe
df_22.head()
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df_22, df_23], axis=0)

# Display the results
combined_df
```

```python
# Save the dataframe to a CSV file
combined_df.to_csv('Resources/citibike.csv', index=False)
```
